```python
#Step3: Combine the Datasets
# Standardize column names first (optional but helpful)
df1.columns = df1.columns.str.lower()
df2.columns = df2.columns.str.lower()
df3.columns = df3.columns.str.lower()
df4.columns = df4.columns.str.lower()
df5.columns = df5.columns.str.lower()

# Combine all datasets
df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)

print(df.shape)
```

```python
#Step4: Clean the Data
# Remove duplicates
df = df.drop_duplicates()

# Check missing values
df.isnull().sum()
```

```python
#Step5: Simulate Athlete-Specific Features
import numpy as np

np.random.seed(42)

n = len(df)

# Athlete-Specific Features
df["practice_hours_per_week"] = np.random.randint(5, 25, n)
df["travel_days_per_month"] = np.random.randint(0, 10, n)
df["fatigue_score"] = np.clip(np.random.normal(5, 2, n), 1, 10)

# Pyschology-Based Features
df["academic_motivation_score"] = np.clip(np.random.normal(7, 1.5, n), 1, 10)
df["athletic_motivation_score"] = np.clip(np.random.normal(7.5, 1.5, n), 1, 10)
df["academic_identity_score"] = np.clip(np.random.normal(6.5, 1.5, n), 1, 10)
df["athletic_identity_score"] = np.clip(np.random.normal(7.5, 1.5, n), 1, 10)
df["self_regulation_score"] = np.clip(np.random.normal(6.5, 1.5, n), 1, 10)
df["role_negotiation_score"] = np.clip(np.random.normal(6.0, 1.5, n), 1, 10)
```

```python
import pandas as pd
import numpy as np

# Step 6: Feature Engineering

# Create fallback columns if missing
if "study_hours_per_week" not in df.columns:
    if "studytime" in df.columns:
        df["study_hours_per_week"] = df["studytime"]
    elif "hours_studied" in df.columns:
        df["study_hours_per_week"] = df["hours_studied"]
    else:
        df["study_hours_per_week"] = 0

if "assignments" not in df.columns:
    df["assignments"] = 0

if "attendance_rate" not in df.columns:
    if "attendance" in df.columns:
        df["attendance_rate"] = df["attendance"]
    else:
        df["attendance_rate"] = 75

if "previous_gpa" not in df.columns:
    if "gpa" in df.columns:
        df["previous_gpa"] = df["gpa"]
    elif "g3" in df.columns:
        df["previous_gpa"] = df["g3"]
    else:
        df["previous_gpa"] = 2.5

# Remove % signs if present, then convert to numeric
for col in [
    "study_hours_per_week",
    "assignments",
    "attendance_rate",
    "previous_gpa",
    "practice_hours_per_week",
    "travel_days_per_month",
    "fatigue_score",
    "academic_motivation_score",
    "self_regulation_score",
    "athletic_identity_score",
    "academic_identity_score"
]:
    if col in df.columns:
        df[col] = df[col].astype(str).str.replace("%", "", regex=False).str.strip()
        df[col] = pd.to_numeric(df[col], errors="coerce")

# Fill missing values after conversion
fill_values = {
    "study_hours_per_week": 0,
    "assignments": 0,
    "attendance_rate": 75,
    "previous_gpa": 2.5,
    "practice_hours_per_week": 10,
    "travel_days_per_month": 2,
    "fatigue_score": 5,
    "academic_motivation_score": 6,
    "self_regulation_score": 6,
    "athletic_identity_score": 6,
    "academic_identity_score": 6
}

for col, val in fill_values.items():
    if col in df.columns:
        df[col] = df[col].fillna(val)

# 1. Academic Load
df["academic_load"] = df["study_hours_per_week"] + df["assignments"]

# 2. Athlete Stress Index
df["athlete_stress_index"] = (
    df["practice_hours_per_week"] * 0.4 +
    df["travel_days_per_month"] * 0.3 +
    df["fatigue_score"] * 0.3
)

# 3. Academic Strength
df["academic_strength"] = (
    df["previous_gpa"] * 0.5 +
    df["academic_motivation_score"] * 0.3 +
    df["self_regulation_score"] * 0.2
)

# 4. Identity Conflict
df["identity_conflict"] = (
    df["athletic_identity_score"] - df["academic_identity_score"]
)

# 5. Engagement Score
df["engagement_score"] = (
    df["attendance_rate"] * 0.4 +
    df["study_hours_per_week"] * 0.3 +
    df["self_regulation_score"] * 0.3
)

print(df[[
    "academic_load",
    "athlete_stress_index",
    "academic_strength",
    "identity_conflict",
    "engagement_score"
]].head())
```

```python
print(df[["attendance_rate", "study_hours_per_week", "self_regulation_score"]].dtypes)
print(df[["attendance_rate", "study_hours_per_week", "self_regulation_score"]].head(10))
```

```python
# Step 7: Create Target Variable (At-Risk)

import numpy as np
import pandas as pd

# Ensure numeric columns
cols = [
    "previous_gpa",
    "attendance_rate",
    "study_hours_per_week",
    "fatigue_score",
    "self_regulation_score",
    "athlete_stress_index",
    "academic_motivation_score"
]

for col in cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="ignore")

# Fill missing values
for col in cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

# Create composite risk score
risk_score = (
    -1.2 * df["previous_gpa"]
    -0.03 * df["attendance_rate"]
    -0.08 * df["study_hours_per_week"]
    +0.20 * df["fatigue_score"]
    -0.20 * df["self_regulation_score"]
    +0.25 * df["athlete_stress_index"]
    -0.15 * df["academic_motivation_score"]
)

# Balanced threshold
threshold = risk_score.median()

df["at_risk"] = (risk_score > threshold).astype(int)

print("Risk threshold:", threshold)
print(df["at_risk"].value_counts())
```
